```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt

order_time=[2,5,3,1,4]
delivery_time=[1.5,3,2,2.5,1]
delivery_partner=["anu","balu","chandu","dolly","smiley"]
cus_rating=[2,3,5,1,4]

for i in range(len(order_time)):
    print(order_time[i],delivery_time[i],delivery_partner[i],cus_rating[i])
#average delivery time
avg_del_time=sum(delivery_time)/len(delivery_time)
print(avg_del_time)
#peak order hours
peak_order_hours=statistics.mode(order_time)
print(f"Peak Order Hours: {peak_order_hours}")
#most delayed areas
most_delayed_areas_list = []
for i in range(len(delivery_time)):
  if delivery_time[i] > avg_del_time:
    most_delayed_areas_list.append(delivery_partner[i])
print(f"Most Delayed Areas: {most_delayed_areas_list}")
#top customer rating
top_customer_rating=sorted(cus_rating, reverse=True)[:3]
print(f"Top Customer Rating: {top_customer_rating}")

#visualization
plt.figure(figsize=(9, 6))
plt.bar(delivery_partner, delivery_time)
plt.xlabel('Delivery Partner')
plt.ylabel('Delivery Time (hours)')
plt.title('\nDelivery Time by Delivery Partner')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data={
    "product_purchases":["shoes","watch","dress","watch","necklace"],
    "cart_abondement":[0,1,1,0,1],
    "customer_ratings":[2,5,1,4,3],
    "payment_methods":["upi","cod","creditcard","upi","cod"]
}
df=pd.DataFrame(data)
print(df.to_string(index=False))
most_soldproducts=df["product_purchases"].mode()
print(f"\nMost Sold Products: {most_soldproducts.to_string(index=False)}")
#high revenue customers
high_revenue_customers=df["customer_ratings"].max()
print(f"\nHigh Revenue Customers: {high_revenue_customers}")
most_paymentmethod=df["payment_methods"].mode()
print(f"\nMost used payment methods : {most_paymentmethod.to_string(index=False)}")
#products with poor rating
poor_rating_products=df[df["customer_ratings"]<3]["product_purchases"]
print(f"\nProducts with poor rating: {poor_rating_products.to_string(index=False)}")
#visualization
plt.figure(figsize=(9, 6))
plt.bar(df["product_purchases"], df["cart_abondement"])
plt.xlabel('Product Purchases')
plt.ylabel('Cart Abondment')
plt.title('\nCart Abondment by Product Purchases')
plt.show()

# Pie chart for product purchases
product_counts = df["product_purchases"].value_counts()
plt.figure(figsize=(9, 6))
plt.pie(product_counts, labels=product_counts.index)
plt.title('\nDistribution of Product Purchases')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {
    "transaction_amount": [1000, 2500, 50.78, 3000, 6000],
    "location": ["hyd", "bng", "ap", "mpl", "tdp"],
    "time": [4, 5, 2, 1, 3],
    "amount_activity": ["active", "inactive", "active", "active", "inactive"]
}
df=pd.DataFrame(data)
print(df.to_string(index=False))
#unusal transactions
df["unusual_transaction"] = df["transaction_amount"] % 1 != 0

print("Unusual Transactions:")

#high risk account
df["high_risk_account"] = df["unusual_transaction"]
high_risk_accounts = df[df["high_risk_account"] == True]

print(high_risk_accounts)
#fraud pattertns
fraud_pattern = df.query(
    "transaction_amount > 5000 or  amount_activity == 'inactive'"
)

print(fraud_pattern)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['transaction_amount'], bins=5)
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Transaction Amounts')
plt.grid(axis='y')
plt.show()
```
