```python
import pandas as pd
import numpy as np
marks=[70,88,90,87,92]
hours=[2,2,3,2,3]

# Create a DataFrame for easier plotting and statistical description
df = pd.DataFrame({
    'Marks': marks,
    'Hours_Studied': hours
})

print("Mean",np.mean(marks))
print("median",np.median(marks))
print("standard deviation",np.std(marks))
print("variance",np.var(marks))
print("corelation",np.corrcoef(marks,hours))
```

### Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.scatterplot(x='Hours_Studied', y='Marks', data=df, s=100)
plt.title('Marks vs. Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

### Detailed Statistical Description

```python
print('--- Statistical Description of Data ---\n')
display(df.describe())
```

### Linear Regression Analysis

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Prepare the data for regression
X = df[['Hours_Studied']]
y = df['Marks']

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the regression line
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Hours_Studied', y='Marks', data=df, s=100, label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression: Marks vs. Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficient (Hours Studied): {model.coef_[0]:.2f}")
print(f"R-squared: {r2_score(y, y_pred):.2f}")
```

### Additional Data Visualizations

```python
# Bar chart for individual Marks
plt.figure(figsize=(8, 5))
sns.barplot(x=df.index, y='Marks', data=df, hue=df.index, palette='viridis', legend=False)
plt.title('Marks per Student')
plt.xlabel('Student Index')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Pie chart for distribution of Hours Studied
hours_counts = df['Hours_Studied'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(hours_counts, labels=hours_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Hours Studied')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# ============================================
# STATISTICAL ANALYSIS IN GOOGLE COLAB
# ============================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
from sklearn.linear_model import LinearRegression

# --------------------------------------------
# Create Dataset
# --------------------------------------------

data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct"],
    "Advertising_Cost": [10,12,15,18,20,22,25,28,30,35],
    "Sales": [100,120,135,150,170,180,200,220,240,260]
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

# --------------------------------------------
# DESCRIPTIVE STATISTICS
# --------------------------------------------

print("\n==============================")
print("DESCRIPTIVE STATISTICS")
print("==============================")

print("\nMean:")
print(df[["Advertising_Cost", "Sales"]].mean())

print("\nMedian:")
print(df[["Advertising_Cost", "Sales"]].median())

print("\nMode:")
print(df[["Advertising_Cost", "Sales"]].mode())

print("\nVariance:")
print(df[["Advertising_Cost", "Sales"]].var())

print("\nStandard Deviation:")
print(df[["Advertising_Cost", "Sales"]].std())

print("\nMinimum Values:")
print(df[["Advertising_Cost", "Sales"]].min())

print("\nMaximum Values:")
print(df[["Advertising_Cost", "Sales"]].max())

print("\nRange:")
print(df[["Advertising_Cost", "Sales"]].max() - df[["Advertising_Cost", "Sales"]].min())

# --------------------------------------------
# CORRELATION
# --------------------------------------------

print("\n==============================")
print("CORRELATION")
print("==============================")

correlation = df["Advertising_Cost"].corr(df["Sales"])

print("\nCorrelation between Advertising Cost and Sales:")
print(correlation)

# --------------------------------------------
# REGRESSION ANALYSIS
# --------------------------------------------

print("\n==============================")
print("LINEAR REGRESSION")
print("==============================")

# Independent Variable (X)
X = df[["Advertising_Cost"]]

# Dependent Variable (Y)
y = df["Sales"]

# Create Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Predictions
predictions = model.predict(X)

# Regression Equation
slope = model.coef_[0]
intercept = model.intercept_

print(f"\nRegression Equation:")
print(f"Sales = {intercept:.2f} + ({slope:.2f} * Advertising_Cost)")

# R-Squared Value
r_squared = model.score(X, y)

print(f"\nR-squared Value: {r_squared:.4f}")

# --------------------------------------------
# BAR CHART
# --------------------------------------------

plt.figure(figsize=(10,5))

x = np.arange(len(df["Month"]))

width = 0.35

plt.bar(x - width/2, df["Advertising_Cost"], width, label='Advertising Cost')
plt.bar(x + width/2, df["Sales"], width, label='Sales')

plt.xticks(x, df["Month"])

plt.xlabel("Month")
plt.ylabel("Values")
plt.title("Advertising Cost and Sales")
plt.legend()

plt.show()

# --------------------------------------------
# PIE CHART
# --------------------------------------------

plt.figure(figsize=(8,8))

plt.pie(
    df["Sales"],
    labels=df["Month"],
    autopct='%1.1f%%'
)

plt.title("Sales Distribution by Month")

plt.show()

# --------------------------------------------
# HISTOGRAM
# --------------------------------------------

plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=5)

plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.title("Histogram of Sales")

plt.show()

# --------------------------------------------
# SCATTER PLOT WITH REGRESSION LINE
# --------------------------------------------

plt.figure(figsize=(8,5))

plt.scatter(df["Advertising_Cost"], df["Sales"])

plt.plot(df["Advertising_Cost"], predictions)

plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.title("Regression Analysis")

plt.show()

# --------------------------------------------
# FULL STATISTICAL DESCRIPTION
# --------------------------------------------

print("\n==============================")
print("FULL STATISTICAL DESCRIPTION")
print("==============================")

print(df.describe())

# --------------------------------------------
# INTERPRETATION
# --------------------------------------------

print("\n==============================")
print("INTERPRETATION")
print("==============================")

print("""
1. Mean gives the average value.

2. Median gives the middle value.

3. Standard deviation shows how spread the data is.

4. Correlation close to +1 means strong positive relationship.

5. Regression shows that as advertising cost increases,
   sales also increase.

6. R-squared close to 1 means the regression model fits well.
""")
```

```python
# ============================================
# HEALTH DATA STATISTICAL ANALYSIS
# GOOGLE COLAB CODE
# ============================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
from sklearn.linear_model import LinearRegression

# --------------------------------------------
# CREATE DATASET
# --------------------------------------------

data = {
    "Person": ["P1","P2","P3","P4","P5","P6","P7","P8","P9","P10"],
    "Exercise_Hours": [1,2,3,4,5,6,2,3,5,6],
    "Weight_kg": [85,82,80,76,72,70,84,78,74,68],
    "Blood_Pressure": [140,138,132,128,124,120,136,130,125,118]
}

df = pd.DataFrame(data)

print("===================================")
print("HEALTH DATASET")
print("===================================")

print(df)

# --------------------------------------------
# DESCRIPTIVE STATISTICS
# --------------------------------------------

print("\n===================================")
print("DESCRIPTIVE STATISTICS")
print("===================================")

print("\nMean:")
print(df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].mean())

print("\nMedian:")
print(df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].median())

print("\nMode:")
print(df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].mode())

print("\nVariance:")
print(df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].var())

print("\nStandard Deviation:")
print(df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].std())

print("\nMinimum Values:")
print(df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].min())

print("\nMaximum Values:")
print(df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].max())

print("\nRange:")
print(df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].max() - df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].min())

# --------------------------------------------
# CORRELATION MATRIX
# --------------------------------------------

print("\n===================================")
print("CORRELATION MATRIX")
print("===================================")

correlation_matrix = df[["Exercise_Hours", "Weight_kg", "Blood_Pressure"]].corr()

print(correlation_matrix)

# --------------------------------------------
# LINEAR REGRESSION
# --------------------------------------------

print("\n===================================")
print("LINEAR REGRESSION")
print("===================================")

# Independent Variable (X)
X = df[["Exercise_Hours"]]

# Dependent Variable (Y)
y = df["Blood_Pressure"]

# Create Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Predictions
predictions = model.predict(X)

# Regression Equation
slope = model.coef_[0]
intercept = model.intercept_

print(f"\nRegression Equation:")
print(f"Blood Pressure = {intercept:.2f} + ({slope:.2f} × Exercise Hours)")

# R-Squared Value
r_squared = model.score(X, y)

print(f"\nR-Squared Value: {r_squared:.4f}")

# --------------------------------------------
# BAR CHART
# --------------------------------------------

plt.figure(figsize=(12,6))

x = np.arange(len(df["Person"]))

width = 0.25

plt.bar(x - width, df["Exercise_Hours"], width, label='Exercise Hours')
plt.bar(x, df["Weight_kg"], width, label='Weight')
plt.bar(x + width, df["Blood_Pressure"], width, label='Blood Pressure')

plt.xticks(x, df["Person"])

plt.xlabel("Persons")
plt.ylabel("Values")
plt.title("Health Data Analysis")
plt.legend()

plt.show()

# --------------------------------------------
# PIE CHART
# --------------------------------------------

plt.figure(figsize=(8,8))

plt.pie(
    df["Blood_Pressure"],
    labels=df["Person"],
    autopct='%1.1f%%'
)

plt.title("Blood Pressure Distribution")

plt.show()

# --------------------------------------------
# HISTOGRAM
# --------------------------------------------

plt.figure(figsize=(8,5))

plt.hist(df["Weight_kg"], bins=5)

plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.title("Histogram of Weight")

plt.show()

# --------------------------------------------
# SCATTER PLOT WITH REGRESSION LINE
# --------------------------------------------

plt.figure(figsize=(8,5))

plt.scatter(df["Exercise_Hours"], df["Blood_Pressure"])

plt.plot(df["Exercise_Hours"], predictions)

plt.xlabel("Exercise Hours")
plt.ylabel("Blood Pressure")
plt.title("Regression Analysis")

plt.show()

# --------------------------------------------
# BOX PLOT
# --------------------------------------------

plt.figure(figsize=(6,5))

plt.boxplot(df["Blood_Pressure"])

plt.ylabel("Blood Pressure")
plt.title("Box Plot of Blood Pressure")

plt.show()

# --------------------------------------------
# LINE CHART
# --------------------------------------------

plt.figure(figsize=(10,5))

plt.plot(df["Person"], df["Weight_kg"], marker='o')
plt.plot(df["Person"], df["Blood_Pressure"], marker='o')

plt.xlabel("Persons")
plt.ylabel("Values")
plt.title("Weight and Blood Pressure Trend")

plt.legend(["Weight", "Blood Pressure"])

plt.show()

# --------------------------------------------
# FULL STATISTICAL DESCRIPTION
# --------------------------------------------

print("\n===================================")
print("FULL STATISTICAL DESCRIPTION")
print("===================================")

print(df.describe())

# --------------------------------------------
# INTERPRETATION
# --------------------------------------------

print("\n===================================")
print("INTERPRETATION")
print("===================================")

print("""
1. Mean gives the average exercise hours,
   weight, and blood pressure.

2. Median gives the middle value.

3. Mode shows the most repeated value.

4. Variance and standard deviation show
   how much the data is spread.

5. Correlation matrix shows relationships
   among exercise, weight, and blood pressure.

6. Regression analysis shows how exercise
   affects blood pressure.

7. Negative regression slope means blood
   pressure decreases as exercise increases.

8. R-squared close to 1 means the model
   fits the data well.
""")
```
